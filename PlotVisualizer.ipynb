```python
class Config:
    @staticmethod
    def get_palette():
        '''Returns a color palette for plotting'''
        return ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']  # Customize the colors as needed

    @staticmethod
    def get_style():
        '''Returns plot style (optional)'''
        return 'seaborn-whitegrid'  # You can use a predefined matplotlib style if needed
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_sir_from_csv(csv_file, size=(6,3), include_fatalities=False, title='S-I-R Plot - Max Infection Range (0.01)'):
    '''Plots S-I-R parameters from a CSV file containing "Healthy", "Infected", "Immune", "In Treatment", and "Dead"
    
    Keyword arguments
    -----------------
    csv_file : str
        Path to the CSV file containing the simulation data.
        
    size : tuple
        Size at which the plot will be initialized (default: (6,3))
        
    include_fatalities : bool
        Whether to plot the fatalities as well (default: False)
        
    title : str
        Title of the plot (default: 'S-I-R plot from CSV')
    '''
    
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_file)
    
    # Get the color palette (assuming `Config.get_palette()` is a valid method)
    palette = Config.get_palette()
    
    # Plot the data
    plt.figure(figsize=size)
    plt.title(title)
    
    plt.plot(df['Healthy'], color=palette[0], label='Healthy')
    plt.plot(df['Infected'], color=palette[1], label='Infected')
    plt.plot(df['Immune'], color=palette[2], label='Immune')
    
    if include_fatalities:
        plt.plot(df['Dead'], color=palette[3], label='Fatalities')
    
    # Add axis labels
    plt.xlabel('Time (Frame)')
    plt.ylabel('Population')
    
    # Add legend
    plt.legend()
    
    # Beautify the plot
    plt.tight_layout()
    
    # Display the plot
    plt.show()
```

```python
plot_sir_from_csv('simulation_report.csv', include_fatalities=True, title="S-I-R Plot - Max Treatment Factor (10)")
```
